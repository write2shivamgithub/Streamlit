```python
import pandas as pd
```

```python
df = pd.read_csv('startup_funding.csv')
df
```

```python
sorted(df['Startup Name'].unique().tolist())
```

```python
df.info()
```

```python
df['Investors Name'] = df['Investors Name'].fillna('Undisclosed')
sorted(df['Investors Name'].unique().tolist())
```

```python
df.drop(columns = ['Remarks'], inplace=True)
```

```python
df
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.columns
```

```python
df.rename(columns = {
    'Date dd/mm/yyyy':'date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subvertical',
    'City  Location':'city',
    'Investors Name':'investors',
    'InvestmentnType':'round',
    'Amount in USD':'amount'
},inplace=True)
```

```python
df.info()
```

```python
df['amount'] = df['amount'].fillna('0')
df['amount'] = df['amount'].str.replace(',','')
df['amount'] = df['amount'].str.replace('undisclosed','0')
df['amount'] = df['amount'].str.replace('unknown','0')
df['amount'] = df['amount'].str.replace('Undisclosed','0')
```

```python
df =df[df['amount'].str.isdigit()]
df
```

```python
df['amount'] = df['amount'].astype('float')
```

```python
df.info()
```

```python
def to_inr(dollar):
    inr = dollar * 85
    return inr/10000000
```

```python
df['amount'] = df['amount'].apply(to_inr)
```

```python
df['date'] = pd.to_datetime(df['date'],errors='coerce') #coerce here means row which are not converting is going to delete automatically
```

```python
df.info()
```

```python
df = df.dropna(subset=['date','startup','vertical','city','investors','round','amount'])
df.info()
```

```python
sorted(set(df['investors'].str.split(',').sum())) #set( helps to return unique value)
```
